## **Seaborn**  
**Seaborn은 Matplotlib에 기반하여 제작된 파이썬 데이터 시각화 모듈**
1. seaborn에서만 제공되는 통계 기반 plot
2. 특별하게 꾸미지 않아도 깔끔하게 구현되는 기본 color
3. 더 아름답게 그래프 구현이 가능한 palette 기능
4. pandas 데이터프레임과 높은 호환성  
: hue 옵션으로 bar 구분이 가능하며, xtick, ytick, xlabel, ylabel, legend 등이 추가적인 코딩 작업없이 자동으로 세팅

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sns
```

```python
x = np.linspace(0, 10, 100) # np.linspace(처음, 끝, 간격)
y1 = np.sin(x)
y2 = 2*np.sin(x+0.5)
y3 = 3*np.sin(x+1.0)
y4 = 4*np.sin(x+1.5)

plt.figure(figsize=(10,6))
plt.plot(x,y1, x,y2, x,y3, x,y4)
plt.show()
```

```python
sns.set_style("white")

plt.figure(figsize=(10,6))
plt.plot(x,y1, x,y2, x,y3, x,y4)

sns.despine()

plt.show()
```

```python
sns.set_style("dark") # 배경 dark

plt.figure(figsize=(10,6))
plt.plot(x,y1, x,y2, x,y3, x,y4)
plt.show()
```

```python
sns.set_style("whitegrid") # grid

plt.figure(figsize=(10,6))
plt.plot(x,y1, x,y2, x,y3, x,y4)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(x,y1, x,y2, x,y3, x,y4)

sns.despine(offset=10) # sns.despine() : 테두리 제거

plt.show()
```

## **example #1**
#### **흡연자와 비흡연자간 음식점 종업원에게 주는 팁에 유의미한 차이가 있는지 확인**

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set_style("whitegrid")
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
tips.head(5)
```

```python
sns.set_style("whitegrid")

plt.figure(figsize=(8,6))
sns.boxplot(x=tips["total_bill"])
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", data=tips) # boxplot 상 x 축과 y 축을 따로따로 지정해 시각화
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", hue="smoker", data=tips, palette="Set3")
plt.show()
```

- hue 라는 단어는 seaborn 에서만 있는 키워드 옵션으로,
- 한 시각화 안에서 차원을 추가하는 역할. (nominal variable을 추가)

```python
plt.figure(figsize=(8,6))
sns.swarmplot(x="day", y="total_bill", data=tips, color=".5")
plt.show()
```

**dot plot이용한 시각화의 경우, 데이터가 몰려있을 때 점 끼리 겹치게 되므로 정보에 대한 설명력이 떨어지는 단점.**  
-> **swarmplot** : 직교벡터에 해당하는 차원을 하나 추가하여 데이터가 몰려있는 경우 직교성분 값을 주어 데이터 쏠림에 의한 설명력 저하 방지

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", data=tips)
sns.swarmplot(x="day", y="total_bill", data=tips, color=".25")
plt.show()
```

**박스플롯은 quantile 과 outlier 정보만 들어있어서, 데이터의 분포를 알 수 있는 swarmplot 결합**

### **Other**

```python
sns.set_style("darkgrid")
sns.lmplot(x="total_bill", y="tip", data=tips, size=7)
plt.show()
```

```python
sns.set_style("darkgrid")
sns.lmplot(x="total_bill", y="tip", data=tips, height=7) # The `size` parameter has been renamed to `height`
plt.show()
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips, palette="Set1", height=7)
plt.show()
```

## **example #3**
#### **flights**

```python
flights = sns.load_dataset("flights")
flights.head(5)
```

```python
flights = flights.pivot("month", "year", "passengers")
flights.head(5)
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(flights)
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(flights, annot=True, fmt="d")
plt.show()
```

**fmt:**  
String formatting code to use when adding annotations.
- d 는 decimal(10진수)

**annot:**  
If True, write the data value in each cell.  
If an array-like with the same shape as data,then use this to annotate the heatmap instead of the raw data.  
- 각 네모의 안에 네모에 해당하는 값을 적어주는 옵션

## **example #4**
#### **iris**

```python
sns.set(style="ticks")
iris = sns.load_dataset("iris")
iris.head(10)
```

```python
sns.pairplot(iris)
plt.show()
```

**sns.pairplot** : 모든 컬럼들에 대해 산점도 행렬을 그려주는 기능  
- 각 컬럼들 간의 상호 상관관계 확인 가능

```python
sns.pairplot(iris, hue="species")
plt.show()
```

#### **데이터 크기가 큰 경우, pairplot 시 시간이 너무 오래 걸림 -> 보고자 하는 컬럼들만 선택**

```python
sns.pairplot(iris, vars=["sepal_width", "sepal_length"])
plt.show()
```

```python
sns.pairplot(iris, x_vars=["sepal_width", "sepal_length"], 
             y_vars=["petal_width", "petal_length"])
plt.show()
```

**x축과 y 축을 각각 지정 가능  
기존에는, 대각행렬 성분이 자기 자신과의 비교이기 때문에 값의 분포를 나타내는 정보**

## **example #5**
#### **anscombe**

```python
anscombe = sns.load_dataset("anscombe")
anscombe.head(5)
```

```python
sns.set_style("darkgrid")
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'I'"),  ci=None, height=7)
# ci(Confidence Interval) = None : 신뢰구간을 의도적으로 지움
plt.show()
```

```python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'I'"),
           ci=None, scatter_kws={"s": 80}, height=7) # scatter_kws : 산점도를 그릴 때 점의 모양과 크기를 넣어주는 옵션
plt.show()
```

```python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'II'"),
           order=1, ci=None, scatter_kws={"s": 80}, height=7)
plt.show()
```

**선형관계가 아닌 curvilinear 관계일 경우를 나타내는 option, 선형회귀를 할 수 없음을 보여준다  
-> 1차 불가 2차로 진행: order=2**

```python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'II'"),
           order=2, ci=None, scatter_kws={"s": 80}, height=7)
plt.show()
```

**order 옵션을 통해 2차 함수로 fitting 한 코드  
일반 1차함수 fitting 이 기본값이었고, 이를 옵션을 통해 2로 바꾼 모습**

```python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'III'"),
           ci=None, scatter_kws={"s": 80}, height=7)
plt.show()
```

**robust 옵션을 이용해서 너무 튀는 데이터는 무시**

```python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'III'"),
           robust=True, ci=None, scatter_kws={"s": 80}, height=7)
plt.show()
```
