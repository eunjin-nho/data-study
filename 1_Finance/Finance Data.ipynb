```python
sectors = dict(list(df_krx.groupby('Sector')))
semi_sec = sectors['반도체 제조업']
semi_sec.head(10)
```

### **02.개별 종목 가져오기**

```python
import pandas as pd
import FinanceDataReader as fdr
```

```python
Semi = pd.DataFrame()

for i, row in semi_sec.iterrows():
    symbol, name = row['Symbol'], row['Name']
    print(symbol, name)
    
    df = fdr.DataReader(symbol, '2020-01-01', '2020-12-31')
    Semi[name] = df['Close'] # 가격 데이터의 종가(Close)를 컬럼으로 추가
```

```python
Semi
```

**결측치 제거**

```python
Semi = Semi.dropna(axis=1)
Semi
```

**하나 사라진 것 확인**

### **03.수익률 계산**

```python
acc_rets = Semi / Semi.iloc[0] - 1.0
acc_rets
```

**최종 수익률은 마지막 행(-1)**
- 수익률 기준 sorting

```python
returns = acc_rets.iloc[-1]
```

```python
returns.sort_values(ascending=False)
```

### **12월 한달동안 비교**

```python
df = Semi['2020-12-01':'2020-12-31']
df
```

```python
acc_rets = df / df.iloc[0] - 1.0
returns = acc_rets.iloc[-1]
returns.sort_values(ascending=False)
```

**'2020-12-31' 시점을 기준으로 과거 5일, 10일, 20일, 60일, 120일, 240일 각각 수익률 check**

```python
from datetime import datetime, timedelta

the_day = datetime(2020, 12, 31)

for days in [5, 10, 20, 60, 120, 240]:
    start = the_day - timedelta(days)
    end = the_day
    print(start, '~', end)
```

```python
from datetime import datetime, timedelta

the_day = datetime(2020, 12, 31)
row_dict = {}
for days in [5, 10, 20, 60, 120, 240]:
    start = the_day - timedelta(days)
    end = the_day

    df = Semi[start:end] # 특정 기간
    acc_rets = df / df.iloc[0] - 1.0
    row_dict[days] = acc_rets.iloc[-1]
```

```python
df_rets = pd.DataFrame(row_dict)
df_rets
```

```python
df_rets.mean() # 기간별 수익률
```

## **Data Frame Style변경**

```python
def color_negative_red(val):
    color = 'red' if val < 0 else 'black' # 마이너스의 경우 빨간색
    return 'color: %s' % color

def highlight_max(s):
    is_max = s == s.max()
    return ['background-color: yellow' if v else '' for v in is_max] # MAX에 노란 하이라이트

df_rets.style.\
    applymap(color_negative_red).\
    apply(highlight_max)
```
